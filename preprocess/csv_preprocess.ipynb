```python
import pandas as pd
air_data = pd.read_csv("../2015_Air_quality_in_northern_Taiwan.csv")
air_data.info()
```

```python
air_data['PH_RAIN']
```

```python
#calcuating how much miss value doew this data have
miss_values_count = air_data.isnull().sum()
miss_values_count
```

```python
#convert object => time
air_data['time'] = pd.to_datetime(air_data['time'])
```

```python
#replace PH_RAIN RAINFALL, RAIN_ COND,  "NR" => 0
air_data.loc[:, ["PH_RAIN",'RAINFALL', 'RAIN_COND' ]] = air_data.loc[:,["PH_RAIN",'RAINFALL', 'RAIN_COND' ]].replace(['NR'],[0])
```

```python
#search all '#,*,x' in the column and remove
for columns in air_data.columns:
    if air_data[columns].dtype == "object":
        air_data[columns] = air_data[columns].str.replace('#','')
        air_data[columns] = air_data[columns].str.replace('\*','')
        air_data[columns] = air_data[columns].str.replace('x','')
```

```python
#replace null to the bfill
air_data = air_data.fillna(method='bfill', axis=0).fillna(0)
air_data
```

```python
#check no '#, *, x' in data
for columns in air_data.columns:
    if air_data[columns].dtype == "object":
        print("="*20)
        print("#num: ", air_data[columns].str.contains('#').sum())
        print("*num: ", air_data[columns].str.contains('\*').sum())
        print("xnum: ", air_data[columns].str.contains('x').sum())
```

```python
#save csv
air_data.to_csv("../pre_2015_Air_quality_in_northern_Taiwan.csv", index=False)
```
